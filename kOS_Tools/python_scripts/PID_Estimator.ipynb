# Calcul de la période d'oscillation avec un Kp limite

```python
import pandas as pd
import plotly.express as px
import numpy as np
from scipy.fft import rfft, rfftfreq
```

### Importer le fichier de données

```python
fichier = "KOS1_0.95.csv"
```

```python
data = pd.read_csv("../" + fichier)
data.head()
```

```python
px.line(x=data.Temps, y=[data.Gaz])
```

### Calculs

```python
shipName = fichier.split("_")[0]
str1 = fichier.split("_")[1]
Ku = float(str1.split(".c")[0])
```

```python
Temps = data.Temps[data.Temps >= 1.5].to_numpy()
Gaz = data.Gaz[data.Temps >= 1.5].to_numpy()
```

```python
step = 0.04
N = len(Temps)
yf = rfft(Gaz)
xf = rfftfreq(N, step)
new_xf, new_yf = xf[xf >= 0.01], np.abs(yf[xf >= 0.01])

px.line(x=new_xf, y=new_yf)
```

```python
freq = new_xf[new_yf == new_yf.max()][0]

Tu = round(1/(freq * 2),2)
```

### Résultats

```python
print(f"Sonde : {shipName}")
print(f"Ku = {Ku}")
print(f"Tu = {Tu}")
print("---------")
print(f"Kp = {round(Ku * 0.6, 2)}")
print(f"Ki = {round(Ku * 1.2 / Tu, 2)}")
print(f"Kd = {round(Ku * 3 * Tu / 40, 2)}")
```
